```python
import pandas as pd

students = pd.read_csv("students_scored.csv")
mentors = pd.read_csv("mentors.csv")

# -----------------------------
# STEP 1: Determine student need
# -----------------------------
def determine_need(row):
    if row["WWS"] < 50 or row["stress_level"] >= 8:
        return "Wellness"
    elif row["APS"] < 60:
        return "Academic"
    elif row["CRS"] < 60:
        return "Career"
    else:
        return "General"

students["Primary_Need"] = students.apply(determine_need, axis=1)

# -----------------------------
# STEP 2: Match mentor
# -----------------------------
def match_mentor(need):
    available = mentors[
        (mentors["expertise"] == need) &
        (mentors["current_load"] < mentors["max_students"])
    ]
    
    if len(available) == 0:
        return "No Mentor Available"
    
    # choose least loaded mentor
    chosen = available.sort_values("current_load").iloc[0]
    return chosen["mentor_name"]

students["Assigned_Mentor"] = students["Primary_Need"].apply(match_mentor)

# -----------------------------
# STEP 3: Intervention type
# -----------------------------
def intervention_type(need):
    if need == "Academic":
        return "Academic Support Plan"
    elif need == "Wellness":
        return "Wellness Counseling"
    elif need == "Career":
        return "Career Guidance Session"
    else:
        return "General Mentoring"

students["Intervention"] = students["Primary_Need"].apply(intervention_type)

# -----------------------------
# STEP 4: Risk alerts
# -----------------------------
def risk_alert(row):
    if row["Category"] == "Red":
        return "URGENT: Immediate Mentor Attention"
    elif row["Category"] == "Yellow":
        return "Monitor Closely"
    else:
        return "Normal"

students["Alert_Level"] = students.apply(risk_alert, axis=1)

# Save final table
students.to_csv("student_mentor_mapping.csv", index=False)

students.head()
```
